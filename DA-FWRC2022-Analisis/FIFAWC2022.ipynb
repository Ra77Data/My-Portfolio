```python
import pandas as pd

df = pd.read_csv('Fifa_world_cup_matches.csv')
```

# Comenzamos limpiando los datos irrelevantes

```python
df.drop(columns = ['date','hour','category','left channel team1','left channel team2', 'right channel team1', 'right channel team2','left inside channel team1', 'left inside channel team2', 'right inside channel team1', 'right inside channel team2'], inplace = True)
df

```

# Transformamos las columnas expresadas en porcentajes(str) en números flotantes(float)

```python
df['possession team1'] = df['possession team1'].str[:-1].astype(float) / 100
df['possession team2'] = df['possession team2'].str[:-1].astype(float) / 100
df['possession in contest'] = df['possession in contest'].str[:-1].astype(float) / 100

df
```

```python
df.info()
```
