### PREPARACIÓN DE LOS DATOS PARA LA MODELIZACIÓN

#### Transformar todas las variables categóricas a númericas

```python
from sklearn.preprocessing import OneHotEncoder

#Categóricas
cat = df_ml.select_dtypes('O')

#Instanciamos
ohe = OneHotEncoder(sparse = False)

#Entrenamos
ohe.fit(cat)

#Aplicamos
cat_ohe = ohe.transform(cat)

#Ponemos los nombres
cat_ohe = pd.DataFrame(cat_ohe, columns = ohe.get_feature_names_out(input_features = cat.columns)).reset_index(drop = True)
```

```python
cat_ohe
```

#### Dataframe final

Seleccionamos las variables numéricas para poder juntarlas a las cat_hoe

```python
num = df.select_dtypes('number').reset_index(drop = True)
```

Las juntamos todas en el dataframe final

```python
df_ml = pd.concat([cat_ohe,num], axis = 1)
df_ml
```

### Cambio de Modelo a Regresión Logistica

```python
x = df_ml.drop(columns='abandono')
y = df_ml['abandono']
```

```python
from sklearn.model_selection import train_test_split

train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 0.3)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score


# Escalar las características (opcional pero recomendado)
scaler = StandardScaler()
scaler.fit(train_x)
train_scaled_x = scaler.transform(train_x)
test_scaled_x = scaler.transform(test_x)

# Crear el modelo de regresión logística
logreg_model = LogisticRegression()

# Entrenar el modelo
logreg_model.fit(train_scaled_x, train_y)

# Predecir las probabilidades en el conjunto de prueba
logreg_predictions = logreg_model.predict_proba(test_scaled_x)[:, 1]

# Calcular la métrica ROC AUC
logreg_auc = roc_auc_score(test_y, logreg_predictions)
print(f"Regresión Logística ROC AUC: {logreg_auc}")
```

```python
met_orig = 0.666
met_upgr = 0.8187
mejora = ((met_upgr / met_orig) - 1) * 100
print('La mejora ha sido del %',round(mejora,2))
```

Si bien la mejor fué sustancial intentaremos mejorarla aún más mediante la utilización de GridSearchCV

```python
from sklearn.model_selection import GridSearchCV

# Define la cuadrícula de hiperparámetros
param_grid = {
    'penalty': ['l1', 'l2'],
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'solver': ['liblinear', 'saga']
}

# Crea el modelo de regresión logística
logreg = LogisticRegression()

# Crea el objeto GridSearchCV
grid_search = GridSearchCV(logreg, param_grid, scoring='roc_auc', cv=5)

# Ajusta el GridSearchCV a los datos de entrenamiento
grid_search.fit(train_scaled_x, train_y)

# Imprime los mejores hiperparámetros encontrados
print(f"Mejores hiperparámetros: {grid_search.best_params_}")

# Entrena el modelo final con los mejores hiperparámetros
best_logreg = grid_search.best_estimator_
```

```python
from sklearn.metrics import roc_auc_score

# Predecir las probabilidades en el conjunto de prueba usando el modelo con los mejores hiperparámetros
pred_proba_y = best_logreg.predict_proba(test_scaled_x)[:, 1]

# Calcular el ROC AUC
roc_auc = roc_auc_score(test_y, pred_proba_y)

# Imprimir el ROC AUC
print(f"ROC AUC del modelo optimizado: {roc_auc}")
```
