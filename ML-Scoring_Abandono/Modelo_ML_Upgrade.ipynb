### CARGA DE LOS DATOS

### PREPARACIÓN DE LOS DATOS PARA LA MODELIZACIÓN

#### Transformar todas las variables categóricas a númericas

```python
from sklearn.preprocessing import OneHotEncoder

#Categóricas
cat = df_ml.select_dtypes('O')

#Instanciamos
ohe = OneHotEncoder(sparse = False)

#Entrenamos
ohe.fit(cat)

#Aplicamos
cat_ohe = ohe.transform(cat)

#Ponemos los nombres
cat_ohe = pd.DataFrame(cat_ohe, columns = ohe.get_feature_names_out(input_features = cat.columns)).reset_index(drop = True)
```

```python
cat_ohe
```

#### Dataframe final

Seleccionamos las variables numéricas para poder juntarlas a las cat_hoe

```python
num = df.select_dtypes('number').reset_index(drop = True)
```

Las juntamos todas en el dataframe final

```python
df_ml = pd.concat([cat_ohe,num], axis = 1)
df_ml
```

### DISEÑO DE LA MODELIZACIÓN

#### Separación predictoras y target

```python
x = df_ml.drop(columns='abandono')
y = df_ml['abandono']
```

#### Separación train y test

```python
from sklearn.model_selection import train_test_split

train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 0.3)
```

### Busqueda de optimizar los hiperparametros para mejorar la metrica del modelo original

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

# Define el espacio de búsqueda de hiperparámetros
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 2, 4, 6, 8, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
}

# Crea el clasificador
dtc = DecisionTreeClassifier()

# Crea el objeto GridSearchCV
grid_search = GridSearchCV(dtc, param_grid, cv=5, scoring='roc_auc')

# Entrena el modelo con los datos de entrenamiento
grid_search.fit(train_x, train_y)

# Imprime los mejores parámetros encontrados
print("Mejores parámetros: ", grid_search.best_params_)

# Imprime el mejor score obtenido
print("Mejor score: ", grid_search.best_score_)

# Usa el modelo optimizado para predicciones
best_dtc = grid_search.best_estimator_
predictions = best_dtc.predict(test_x)
```
